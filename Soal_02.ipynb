Visualisasi nilai mahasiswa

```python
import matplotlib.pyplot as plt
nilai = []
```

Input 10 nilai mahasiswa

```python
for i in range(10):
  n = int(input("Masukkan nilai mahasiswa ke-" + str(i+1) + ": "))
  nilai.append(n)
```

Nilai tertinggi dan terendah

```python
nilai_tertinggi = max(nilai)
nilai_terendah = min(nilai)
```

Hitung jumlah mahasiswa yang lulus

```python
lulus = 0
for n in nilai:
  if n >= 60:
    lulus +=1

tidak_lulus = len(nilai) - lulus
```

Grafik #1: Nilai tertinggi Vs nilai terendah

```python
plt.bar(["Tertinggi", "Terendah"], [nilai_tertinggi, nilai_terendah])
plt.title(["Nilai tertinggi vs terendah"])
plt.ylabel(["Nilai"])
plt.show()
```

Grafik #2: Kelulusan

```python
plt.bar(["Lulus", "Tidak Lulus"], [lulus, tidak_lulus])
plt.title(["Statistik Kelulusan"])
plt.ylabel(["Jumlah"])
plt.show()
```
